1. [Association Rule Generation from Transaction Data]

```python

import os
import warnings
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings("ignore")

path = r'/content/drive/MyDrive/Grocery_Items_12.csv'
grocery_data = pd.read_csv(path)

list = [line.dropna().tolist() for _, line in grocery_data.iterrows()]

te = TransactionEncoder()
te_ary = te.fit(list).transform(list)
final_gro = pd.DataFrame(te_ary, columns=te.columns_)

def gen_assoc_rules(df, min_support, min_confidence):
    frequent_items = apriori(df, min_support=min_support, use_colnames=True)
    ar = association_rules(frequent_items, metric="confidence", min_threshold=min_confidence, num_itemsets=len(frequent_items))
    return ar

unique_items = len(te.columns_)
total_records = len(grocery_data)
most_popular_item = grocery_data.stack().value_counts().idxmax()
most_popular_item_count = grocery_data.stack().value_counts().max()

print(f"Unique items: {unique_items}")
print(f"Total records: {total_records}")
print(f"Most popular item: {most_popular_item}, Transactions: {most_popular_item_count}")

rules = gen_assoc_rules(final_gro, 0.01, 0.08)
print(f"\nAssociation Rules (Support: 0.01, Confidence: 0.08):\n{rules}")

min_support_val = [0.001, 0.005, 0.01]
min_confidence_val = [0.05, 0.075, 0.1]
heatmap_data = []

for msv in min_support_val:
    row = []
    for mct in min_confidence_val:
        rule_count = len(gen_assoc_rules(final_gro, msv, mct))
        row.append(rule_count)
    heatmap_data.append(row)

# Create the heatmap
heatmap_df = pd.DataFrame(heatmap_data, columns=min_confidence_val, index=min_support_val)
sns.heatmap(heatmap_df, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Association Rules Heatmap")
plt.xlabel("Min Confidence Threshold")
plt.ylabel("Min Support Values")
plt.show()
```

3. [Text Classification by fine-tuning LLM model]
> Add blockquote



```python
import json
import torch
from torch.utils.data import Dataset
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
import matplotlib.pyplot as plt
import numpy as np

file_path = "/content/drive/MyDrive/student_12"

def load_data(file_path):
    with open(file_path, 'r') as f:
        return [json.loads(line) for line in f]

train_data = load_data(f"{file_path}/train.json")
test_data = load_data(f"{file_path}/test.json")
val_data = load_data(f"{file_path}/validation.json")

class MultiLabelDataset(Dataset):
    def __init__(self, data, tokenizer, max_length=128):
        self.texts = [item.get('text', '') for item in data]
        self.labels = [item.get('labels', []) for item in data]
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = self.texts[idx]
        labels = torch.tensor(self.labels[idx] if self.labels[idx] else [0] * 11, dtype=torch.float)
        encoding = self.tokenizer(
            text, max_length=self.max_length, padding="max_length", truncation=True, return_tensors="pt"
        )
        return {
            "input_ids": encoding["input_ids"].squeeze(),
            "attention_mask": encoding["attention_mask"].squeeze(),
            "labels": labels
        }

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
train_dataset = MultiLabelDataset(train_data, tokenizer)
val_dataset = MultiLabelDataset(val_data, tokenizer)
test_dataset = MultiLabelDataset(test_data, tokenizer)

model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=11)

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    logging_dir="./logs",
    logging_steps=10,
    num_train_epochs=5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    learning_rate=2e-5,
    weight_decay=0.01,
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss"
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.where(logits > 0.5, 1, 0)
    accuracy_all = np.mean((preds == labels).all(axis=1))
    accuracy_one = np.mean((preds * labels).sum(axis=1) > 0)
    return {"accuracy_all": accuracy_all, "accuracy_one": accuracy_one}

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)

trainer.train()

train_loss = trainer.state.log_history
num_epochs = training_args.num_train_epochs

epochs = [*range(1, num_epochs + 1)]

train_losses_per_epoch = []
for epoch in range(1, num_epochs + 1):
    epoch_losses = [x['loss'] for x in train_loss if 'loss' in x and x['epoch'] == epoch]
    if epoch_losses:
        avg_loss = sum(epoch_losses) / len(epoch_losses)
        train_losses_per_epoch.append(avg_loss)
    else:
        train_losses_per_epoch.append(0)

val_losses = [x['eval_loss'] for x in train_loss if 'eval_loss' in x]

plt.plot(epochs, train_losses_per_epoch, label="Training Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.show()

test_results = trainer.evaluate(test_dataset)
print("Test Accuracy (All Labels Match):", test_results["eval_accuracy_all"])
print("Test Accuracy (At Least One Label Matches):", test_results["eval_accuracy_one"])
```

2. [Image Classification using CNN]

```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical

img_height = 128
img_width = 128

dir1 = r'/content/drive/MyDrive/cropped/n02086910-papillon'
dir2 = r'/content/drive/MyDrive/cropped/n02088632-bluetick'
dir3 = r'/content/drive/MyDrive/cropped/n02096437-Dandie_Dinmont'
dir4 = r'/content/drive/MyDrive/cropped/n02113023-Pembroke'

def plot_learning_curves(history, model_name):
    train_acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    epochs = range(1, len(train_acc) + 1)
    plt.plot(epochs, train_acc, label='Training accuracy')
    plt.plot(epochs, val_acc, label='Validation accuracy')
    plt.title(f'{model_name}: Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()

def load_images_and_labels(directory):
    images = []
    labels = []
    for filename in os.listdir(directory):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            img = load_img(os.path.join(directory, filename), target_size=(img_height, img_width))
            img_array = img_to_array(img)
            images.append(img_array)
            if directory == dir1:
                labels.append(0)
            elif directory == dir2:
                labels.append(1)
            elif directory == dir3:
                labels.append(2)
            elif directory == dir4:
                labels.append(3)
    return images, labels

class1_img, class1_l = load_images_and_labels(dir1)
class2_img, class2_l = load_images_and_labels(dir2)
class3_img, class3_l = load_images_and_labels(dir3)
class4_img, class4_l = load_images_and_labels(dir4)

images = np.concatenate([class1_img, class2_img, class3_img, class4_img], axis=0)
labels = np.concatenate([class1_l, class2_l, class3_l, class4_l], axis=0)

labels = to_categorical(labels)

X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

X_train = X_train / 255.0
X_val = X_val / 255.0

def create_model(filters_second_layer, model_name):
    model = Sequential([
        Conv2D(8, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
        MaxPooling2D((2, 2)),
        Conv2D(filters_second_layer, (3, 3), activation='relu'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(8, activation='relu'),
        Dense(4, activation='softmax')
    ])
    model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    print(f"Model Summary for {model_name}:")
    model.summary()
    return model

print("Training Base Model with 4 Filters in 2nd Conv Layer...")
base_model = create_model(4, "Base Model")
base_history = base_model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), verbose=2)
plot_learning_curves(base_history, "Base Model")

print("Training Model with 8 Filters in 2nd Conv Layer...")
model_8_filters = create_model(8, "8 Filters Model")
history_8_filters = model_8_filters.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), verbose=2)
plot_learning_curves(history_8_filters, "8 Filters Model")

print("Training Model with 16 Filters in 2nd Conv Layer...")
model_16_filters = create_model(16, "16 Filters Model")
history_16_filters = model_16_filters.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val), verbose=2)
plot_learning_curves(history_16_filters, "16 Filters Model")

```

```python
'''Best Model: The model with 8 filters in the second convolutional layer performed the best, achieving higher validation accuracy and better generalization.
Overfitting/Underfitting:
The base model showed slight underfitting.
The 8-filter model struck a good balance between underfitting and overfitting.
The 16-filter model suffered from severe underfitting, likely due to optimization challenges or insufficient data.'''
```
